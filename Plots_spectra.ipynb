This notebook provides interactive visualisations of Beta Pictoris spectra obtained on **April 29** and **September 10, 2025**, with the Hubble Space Telescope (STIS and COS; program #17790) and the HARPS spectrograph at the 3.6m ESO telescope in La Silla, Chile (program 115.282N).

Both observations reveal absorption signatures from transiting exocomet gaseous tails, detected in lines of metallic ions (Fe II, Ni II, ...) and volatile tracers (C I). This notebook is designed to help explore and visualise these signatures interactively.

# View of the fitted model

Spectra from each epoch were fitted using the model described in Vrignaud & Lecavelier 2026 (A&A, 707, A60) and Vrignaud et al. (in prep., 2026). In this model, each low-velocity exocomet (LVC) is described using few parameters:
- Its line profile (reconstructed using a PCHIP interpolation)
- Its distance to the star (which sets the efficiency of radiative excitation, and thus the excitation state of the gas)
- Its composition (one column density for each species; here we fitted Fe II, Ni II, Si II, Cr II, Mn II, Ca II, S I and C I lines)

For each epoch, several absorption components were modeled: 
- The LVCs (LVC 1, 2 and 3 for April 29 spectrum; LVC 4 and 5 for September 10 spectrum);
- The circumstellar disc (narrow feature at 0 km/s, detected mostly in lines from the ground states of the studied species)
- The ISM, at -10 km/s. 

In total, the model is described by ~ 100 parameters, and is fitted to the entire observed spectrum (10,000 to 20,000 degrees of freedom).
